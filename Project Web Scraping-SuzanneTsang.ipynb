```python
import pandas as pd
from urllib.parse import urlparse
import numpy as np
import nltk.data
from requests_html import HTMLSession
```

```python
def parse_elements(url,element):
    session = HTMLSession()
    r = session.get(url)
    elements = r.html.find(element)
    return elements
```

The function defines the positive words and negative words that are used for emotion analysis. Additionally, it determines the emotion expressed in a given paragraph based on the presence of the specified positive and negative words.

```python
def emotion_detection(content):
    positive_words = ('confidence','trust','exceptional','unique','reward','fun','engage','memorable')
    negative_words = ('doubtful','fear','overwhelmed','nervousness','frustrated','chore','remorse')

    positive = any(content.count(i) > 0 for i in positive_words)
    negative = any(content.count(i) > 0 for i in negative_words)

    if positive == negative == False:
        return "none"
    elif positive != negative:
        return "positive" if positive else "negative"
    else:
        return "mixed"
```

This code takes paragraphs, extracts the text content, and stores it in a dictionary with the key 'Paragraphs'. The resulting dictionary, 'p_dictionary', can be used to access the text content of the paragraphs by using the key 'Paragraphs'.

```python
p_dictionary = {}
p_list = []
for p in range(0, len(paragraphs)):
      p_list.append(paragraphs[p].text)
p_dictionary['Paragraphs'] = p_list
```

```python
p_dictionary['Paragraphs']
```

```python
#Group each paragraph as a row in a table

df = pd.DataFrame.from_dict(p_dictionary)
df
```

```python
medium_df = pd.DataFrame(p_dictionary)
```

```python
medium_df['Emotion'] = medium_df['Paragraphs'].apply(emotion_detection)
```

```python
# Show the emotion nature belongs to that paragraph

medium_df[['Paragraphs', 'Emotion']]
```

```python
# To count the number of emotions by emotion categories

medium_df['Emotion'].value_counts()
```

Display the data in a bar chart format

```python
import matplotlib.pyplot as plt
plt.bar(medium_df['Emotion'].value_counts().index, medium_df['Emotion'].value_counts().values)
plt.xlabel('Emotion')
plt.ylabel('Count')
plt.title('Emotion Analysis of Medium Post')
plt.show()
```
